## 전체 데이터 수집

```python
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

def safe_select_text(soup, selector, index=None):
    try:
        if index is not None:
            return soup.select(selector)[index].text.strip()
        else:
            element = soup.select_one(selector)
            return element.text.strip() if element else None
    except Exception:
        return None

def extract_reviews(soup, review_data):
    meta = {
        "제품명": safe_select_text(soup, "#Contents .prd_name"),
        "브랜드": safe_select_text(soup, "#Contents .prd_brand a"),
        "원가": safe_select_text(soup, "#Contents .price-1"),
        "판매가": safe_select_text(soup, "#Contents .price-2"),
        "리뷰수": safe_select_text(soup, ".product_rating_area .total em"),
        "평균별점": safe_select_text(soup, ".product_rating_area .num strong"),
        "5점비율": safe_select_text(soup, ".graph_list .per", 0),
        "4점비율": safe_select_text(soup, ".graph_list .per", 1),
        "3점비율": safe_select_text(soup, ".graph_list .per", 2),
        "2점비율": safe_select_text(soup, ".graph_list .per", 3),
        "1점비율": safe_select_text(soup, ".graph_list .per", 4),
        "세정력/아주만족해요": safe_select_text(soup, ".poll_all .list .per", 0),
        "세정력/보통이에요": safe_select_text(soup, ".poll_all .list .per", 1),
        "세정력/다소아쉬워요": safe_select_text(soup, ".poll_all .list .per", 2),
        "자극도/자극없이순해요": safe_select_text(soup, ".poll_all .list .per", 3),
        "자극도/보통이에요": safe_select_text(soup, ".poll_all .list .per", 4),
        "자극도/자극이느껴져요": safe_select_text(soup, ".poll_all .list .per", 5),
        "거품/거품이풍성해요": safe_select_text(soup, ".poll_all .list .per", 6),
        "거품/보통이에요": safe_select_text(soup, ".poll_all .list .per", 7),
        "거품/거품이적어요": safe_select_text(soup, ".poll_all .list .per", 8),
        "향/아주만족해요": safe_select_text(soup, ".poll_all .list .per", 9),
        "향/보통이에요": safe_select_text(soup, ".poll_all .list .per", 10),
        "향/다소아쉬워요": safe_select_text(soup, ".poll_all .list .per", 11)
    }

    reviews = soup.select("#gdasList > li")
    
    if not reviews:
        return 
    
    for review in reviews:
        aspects = review.select(".review_cont .poll_type1 .txt")
        review_data.append({
            **meta,
            "작성자": safe_select_text(review, "p.info_user .id"),
            "작성일자": safe_select_text(review, ".review_cont .date"),
            "리뷰": safe_select_text(review, ".txt_inner"),
            "별점": safe_select_text(review, ".review_point span"),
            "사용현황": safe_select_text(review, "div.badge .point_flag"),
            "세정력": aspects[0].text.strip() if len(aspects) > 0 else None,
            "자극도": aspects[1].text.strip() if len(aspects) > 1 else None,
            "거품": aspects[2].text.strip() if len(aspects) > 2 else None,
            "향": aspects[3].text.strip() if len(aspects) > 3 else None,
        })

for goodsno in range(560):
    print(f"\n상품 {goodsno} 수집 시작")

    max_retries = 5
    retry_count = 0
    driver = None

    while retry_count < max_retries:
        driver = uc.Chrome()
        url = f"https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo={goodsNos[goodsno]}"
        driver.get(url)
        driver.implicitly_wait(5)
        time.sleep(random.uniform(1.5, 2.5)) 

        try:
            driver.find_element(By.CSS_SELECTOR, "#reviewInfo a").click()
            time.sleep(random.uniform(2.0, 3.0))
            break
        except Exception as e:
            print(f"리뷰 탭 클릭 실패 (시도 {retry_count+1}) - {e}")
            driver.quit()
            retry_count += 1
            if retry_count >= max_retries:
                print(f"리뷰 탭 최종 실패, 상품 {goodsno} 건너뜀")
                driver = None
                break

    if not driver:
        continue

    review_data = []
    soup = BeautifulSoup(driver.page_source, "html.parser")
    extract_reviews(soup, review_data)
    print("1페이지 수집 완료")

    current_page = 2
    while True:
        try:
            time.sleep(random.uniform(1.0, 2.0))
            page_links = driver.find_elements(By.CSS_SELECTOR, ".pageing a[data-page-no]")

            clicked = False
            for link in page_links:
                page_no = int(link.get_attribute("data-page-no"))
                if page_no == current_page:
                    driver.execute_script("arguments[0].click();", link)
                    time.sleep(random.uniform(1.5, 2.5))
                    soup = BeautifulSoup(driver.page_source, "html.parser")
                    extract_reviews(soup, review_data)
                    print(f"{current_page}페이지 수집 완료")
                    current_page += 1
                    clicked = True
                    break

            if not clicked:
                next_button = driver.find_elements(By.CSS_SELECTOR, ".pageing a.next")
                if next_button:
                    driver.execute_script("arguments[0].click();", next_button[0])
                    time.sleep(random.uniform(1.5, 2.5))
                else:
                    print("다음 페이지 없음")
                    break
        except Exception as e:
            print(f"오류 발생 - {current_page}페이지: {e}")
            break

    driver.quit()

    if review_data:
        df = pd.DataFrame(review_data)
        df.to_csv(f"./shampoo/oliveyoung_product_{goodsno}.csv", index=False, encoding="utf-8-sig")
        print(f"상품 {goodsno} 리뷰 {len(df)}개 저장 완료")
    else:
        print(f"상품 {goodsno} 리뷰 없음")
```

```python
import pandas as pd
import glob
import os
import re

folder_path = "./data/shampoo"
file_list = glob.glob(os.path.join(folder_path, "oliveyoung_product_*.csv"))

def extract_number(filename):
    match = re.search(r"oliveyoung_product_(\d+).csv", filename)
    return int(match.group(1)) if match else -1

file_list_sorted = sorted(file_list, key=extract_number)

df_list = []
for file in file_list_sorted:
    try:
        df = pd.read_csv(file)
        df_list.append(df)
    except Exception as e:
        print(f"파일 {file} 읽는 중 에러: {e}")

merged_df = pd.concat(df_list, ignore_index=True)
merged_df.to_csv(os.path.join(folder_path, "oliveyoung_all_merged.csv"), index=False)

print("oliveyoung_all_merged.csv 저장 완료")
```

```python
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
import requests
import os

existing_df = pd.read_csv("./oliveyoung_product_info.csv")
existing_goodsNos = set(existing_df["상품번호"].astype(str)) 

all_goodsNos = set(goodsNos) 
missing_goodsNos = list(all_goodsNos - existing_goodsNos) 

new_product_data = []

for idx, goodsno in enumerate(missing_goodsNos):
    print(f"\n[{idx+1}/{len(missing_goodsNos)}] 누락 상품번호 {goodsno} 수집 중...")

    try:
        driver = uc.Chrome()
        url = f"https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo={goodsno}"
        driver.get(url)
        driver.implicitly_wait(5)
        time.sleep(random.uniform(1.5, 2.5))

        soup = BeautifulSoup(driver.page_source, "html.parser")

        product_name = safe_select_text(soup, "#Contents .prd_name")
        image_url = safe_select_attr(soup, "#mainImg", "src")

        if product_name and image_url:
            new_product_data.append({
                "상품번호": goodsno,
                "상품명": product_name,
                "이미지URL": image_url,
                "링크": url
            })
            print(f"수집 성공: {product_name}")
        else:
            print("상품 정보 누락됨")

        driver.quit()
        time.sleep(random.uniform(1.0, 2.0))

    except Exception as e:
        print(f"상품 {goodsno} 수집 실패: {e}")
        try:
            driver.quit()
        except:
            pass
        continue

if new_product_data:
    new_df = pd.DataFrame(new_product_data)
    combined_df = pd.concat([existing_df, new_df], ignore_index=True)
    combined_df.to_csv("./oliveyoung_product_info.csv", index=False, encoding="utf-8-sig")
    print(f"\n총 {len(combined_df)}개 상품 정보 저장 완료 (누락 상품 {len(new_df)}개 추가됨)")
else:
    print("추가로 저장할 상품 정보가 없습니다.")
```
